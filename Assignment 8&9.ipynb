```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
titanic = pd.read_csv('titanic.csv')
titanic
```

```python
titanic.describe(include='all')
```

```python
titanic.isnull().sum()
```

```python
titanic['Embarked'].fillna('S', inplace=True)
```

```python
titanic.describe(include='all')
```

```python
titanic.isnull().sum()
```

```python
titanic['Numberic_Cabin'] = np.where(titanic['Cabin'].isnull(),0,1)
```

```python
titanic.head()
```

```python
mean = titanic['Age'].mean()
mean
```

```python
std = titanic['Age'].std()
std
```

```python
lower = mean-std
lower
```

```python
upper = mean+std
upper
```

```python
titanic['Age'].fillna(np.random.randint(lower,upper),inplace=True)
```

```python
titanic.isnull().sum()
```

```python
titanic['Family'] = titanic['Survived'] + titanic['SibSp']
```

```python
titanic.head()
```

```python
titanic['Is_Alone'] = np.where(titanic['Family']==1,1,0)
```

```python
titanic.head()
```

```python
titanic['Fare_Category'] = pd.cut(titanic['Fare'], bins=4, labels=False) + 1
```

```python
titanic['Fare_Category'].unique()
```

```python
titanic.drop('Cabin',axis=1,inplace=True)
```

```python
titanic
```

```python
surv = pd.DataFrame(titanic.groupby('Survived')['PassengerId'].count())
```

```python
surv.rename(columns={'PassengerId' : 'No of Passengers'},inplace=True)
```

```python
surv
```

```python
import matplotlib.pyplot as plt
```

```python
surv.dtypes
```

```python
labels = surv.index
plt.pie(surv['No of Passengers'], labels=labels,autopct='%1.0f%%')
plt.title('Number of People Survived')
plt.legend()
plt.show()
```
